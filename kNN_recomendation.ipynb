```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./hardwareStore.csv")
```

```python
df
```

```python
X = df[['PRODUCT_ID','STANDARD_COST', 'LIST_PRICE', 'WAREHOUSE_ID']]
y = df['CATEGORY_ID']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
distances, indices = knn.kneighbors(X=df2, n_neighbors=5, return_distance=True)
indices[0]
```

```python
df.iloc[indices[0]]
```

```python
product = []
distance = []

for i in range(0, len(distances.flatten())):
    if i != 0:
        product.append(df.index[indices.flatten()[i]])
        distance.append(distances.flatten()[i])    

p=pd.Series(product,name='product')
d=pd.Series(distance,name='distance')
recommend = pd.concat([p,d], axis=1)
recommend = recommend.sort_values('distance',ascending=True)

print('Recommendations:')
for i in range(0,recommend.shape[0]):
    print('{0}: {1}, with distance of {2}'.format(i, recommend["product"].iloc[i], recommend["distance"].iloc[i]))
```
